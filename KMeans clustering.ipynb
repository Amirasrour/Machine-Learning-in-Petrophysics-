```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
```

```python
df = pd.read_csv('Data_force2020_data_unsupervised_learning.csv', index_col='DEPTH_MD')
```

Data Cleaning

```python
df.dropna(inplace= True)
```

Transform the data 
standarise the data using the StandardScaler
to acccount for variations in measurements units and scale.

```python
scaler = StandardScaler()
```

```python
df[['RHOB_T','NPHI_T','GR_T','PEF_T','DTC_T']]= scaler.fit_transform(df[['RHOB','NPHI','GR','PEF','DTC']])
```

```python
df
```

clustring the data

```python
def optimise_k_means(data, Max_K):
    means = []
    inertias = []
    for K in range(1,Max_K):
        Kmeans = KMeans(n_clusters=K)
        Kmeans.fit(data)
        means.append(K)
        inertias.append(Kmeans.inertia_)
        
#creat the elbow plot
    fig = plt.subplots(figsize=(10,5))
    plt.plot(means, inertias, 'o-')
    plt.xlabel('Number of clusters')
    plt.ylabel('Inertia')
    plt.grid(True)
    plt.show()
   
   
```

```python
optimise_k_means(df[['RHOB_T','NPHI_T']], 10)
```

Applying KMeans clustering

```python
KMeans = KMeans(n_clusters=3)
```

```python
KMeans.fit(df[['NPHI_T','RHOB_T']])
```

```python
df['KMeans_3'] = KMeans.labels_
```

```python
df
```

ploting the results

```python
plt.scatter(x=df['NPHI'], y=df['RHOB'], c=df['KMeans_3'])
plt.ylim(3, 1.5)
plt.xlim(-0.1, 1)
plt.show()
```
